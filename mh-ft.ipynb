#### Dataset Init

````python
from datasets import Dataset, load_dataset

if BUILD_DATASET:
    ds = Dataset.from_list(messages_dataset).train_test_split(test_size=0.5)
    ds.push_to_hub(DATASET_NAME, private=True)
else:
    ds = load_dataset(path=f"matthewhaynesonline/{DATASET_NAME}")
````

### Base Model

````python
from transformers import AutoModelForCausalLM, AutoTokenizer
from trl import SFTConfig, SFTTrainer, setup_chat_format

from notebook_utils import get_torch_device 

device = get_torch_device()

BASE_MODEL_OWNER = "HuggingFaceTB"

# BASE_MODEL_NAME = "SmolLM2-135M"
BASE_MODEL_NAME = "SmolLM2-1.7B"

USE_INSTRUCT_BASE_MODEL = True

if USE_INSTRUCT_BASE_MODEL:
  BASE_MODEL_NAME = f"{BASE_MODEL_NAME}-Instruct"

FULL_BASE_MODEL_NAME = f"{BASE_MODEL_OWNER}/{BASE_MODEL_NAME}"

model = AutoModelForCausalLM.from_pretrained(
    pretrained_model_name_or_path=FULL_BASE_MODEL_NAME
).to(device)

tokenizer = AutoTokenizer.from_pretrained(pretrained_model_name_or_path=FULL_BASE_MODEL_NAME)

if not USE_INSTRUCT_BASE_MODEL:
    model, tokenizer = setup_chat_format(model=model, tokenizer=tokenizer)
````

### Finetune Config

````python
FINETUNE_NAME = f"{BASE_MODEL_NAME}-Matt-Shill"
FINETUNE_TAGS = ["smol", "sft", "matthewhaynes"]
FT_MODEL_DIR = f"{MODELS_OUTPUT_DIR}/{FINETUNE_NAME}"
````

### Trainer Config

````python
sft_config = SFTConfig(
    output_dir=f"{MODELS_OUTPUT_DIR}/sft_output",
    max_steps=1000,  # Adjust based on dataset size and desired training duration
    per_device_train_batch_size=4,  # Set according to your GPU memory capacity
    learning_rate=5e-5,  # Common starting point for fine-tuning
    logging_steps=10,  # Frequency of logging training metrics
    save_steps=100,  # Frequency of saving model checkpoints
    eval_strategy="steps",  # Evaluate the model at regular intervals
    eval_steps=50,  # Frequency of evaluation
    hub_model_id=FINETUNE_NAME,  # Set a unique name for your model
)

trainer = SFTTrainer(
    model=model,
    args=sft_config,
    train_dataset=ds["train"],
    processing_class=tokenizer,
    eval_dataset=ds["test"],
)
````

### Train

````python
trainer.train()
trainer.save_model(FT_MODEL_DIR)
````

### Eval

````python
prompt = "Should I hire Matt?"
messages = [{"role": "user", "content": prompt}]
````

````python
base_model = AutoModelForCausalLM.from_pretrained(
    pretrained_model_name_or_path=FULL_BASE_MODEL_NAME
).to(device)

base_tokenizer = AutoTokenizer.from_pretrained(pretrained_model_name_or_path=FULL_BASE_MODEL_NAME)

if not USE_INSTRUCT_BASE_MODEL:
    base_model, base_tokenizer = setup_chat_format(model=model, tokenizer=tokenizer)

formatted_prompt = base_tokenizer.apply_chat_template(messages, tokenize=False)

inputs = base_tokenizer(formatted_prompt, return_tensors="pt").to(device)
outputs = base_model.generate(**inputs, max_new_tokens=100)

print("Before training:")
print(base_tokenizer.decode(outputs[0], skip_special_tokens=True))
````

````python
ft_model = AutoModelForCausalLM.from_pretrained(
    pretrained_model_name_or_path=FT_MODEL_DIR
).to(device)

ft_tokenizer = AutoTokenizer.from_pretrained(pretrained_model_name_or_path=FT_MODEL_DIR)

formatted_prompt = ft_tokenizer.apply_chat_template(messages, tokenize=False)

inputs = ft_tokenizer(formatted_prompt, return_tensors="pt").to(device)
outputs = ft_model.generate(**inputs, max_new_tokens=100)

print("After training:")
print(ft_tokenizer.decode(outputs[0], skip_special_tokens=True))
````

### GGUF & Quantize

- [DIY converting safetensors format to GGUF on a Mac ](https://www.reddit.com/r/LocalLLaMA/comments/18elm98/diy_converting_safetensors_format_to_gguf_on_a_mac/)
- [Convert and Upload Your GGUF Model to Hugging Face â€“ Step-by-Step Guide](https://gpustack.ai/convert-and-upload-your-gguf-model-to-huggingface-step-by-step-guide/)

#### Clone and Build `llama.cpp`:

**Install cmake to be able to compile llama.cpp**
```bash
brew install cmake
```

**Compile llama.cpp**
```bash
git clone https://github.com/ggerganov/llama.cpp
cd llama.cpp
cmake -B build
cmake --build build --config Release
```

**Setup llama.cpp python scripts (have to use python 3.11 as of 2025-01-26)**
```bash
python3.11 -m venv llamacpp_env
source llamacpp_env/bin/activate
python3.11 -m pip install -r requirements.txt
```

#### Convert SafeTensors to GGUF:

Use `convert_hf_to_gguf.py` from `llama.cpp` to convert the model:

```bash
python3.11 ./convert_hf_to_gguf.py /path/to/hf_model --outfile /path/to/hf_model.gguf --outtype f16
```

#### (Optional) Quantize the Model:

```bash
./build/bin/llama-quantize /path/to/hf_model.gguf /path/to/hf_model-q6_k.gguf Q6_K
```
